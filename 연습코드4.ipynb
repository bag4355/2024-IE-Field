```python
emotion_dict = {
    '행복': 5,       # 매우 긍정적
    '추천': 4,       # 긍정적
    '좋': 4,         # 긍정적
    '아쉬움': 3,     # 중립적
    '심심': 2,       # 부정적
    '안전': 4,       # 긍정적
    '재미': 2,       # 부정적
    '혼자': 2,       # 부정적
    '춥': 2,         # 부정적
    '좋': 4,         # 긍정적
    '행복': 5,       # 매우 긍정적
    '위': 3,         # 중립적
    '편리': 4,       # 긍정적
    '못하': 2,       # 부정적
    '즐기': 4,       # 긍정적
}

# 토큰에 감정 표현 점수 매기기
def label_emotional_tokens(tokens):
    token_labels = {}
    for token in tokens:
        if token in emotion_dict:
            token_labels[token] = emotion_dict[token]
        else:
            token_labels[token] = 0  # 감정 표현이 아닌 경우, 점수 없음
    return token_labels

# 레이블링 결과
token_labels = label_emotional_tokens(tokens)
print("토큰 레이블링 결과:", token_labels)
```

```python
# 평가 항목 정의
evaluation_items = {
    '시설': ['볼링장', '포켓볼', '피씨방', '헬스장', '수영장', '클라이밍벽', '스쿼시', '농구장', '요가', '테니스코트'],
    '위치': ['캠퍼스', '기숙사', '해변', 'la', '플러턴', '캘리포니아'],
    '안전': ['캠퍼스폴리스', '안전'],
    '친구 관계': ['친구', '기숙사', '친구사귀다'],
    '날씨': ['날씨', '춥', '비', '행복']
}

# 평가 항목 추출
def extract_evaluation_items(tokens):
    item_counts = {key: 0 for key in evaluation_items.keys()}
    for token in tokens:
        for item_category, item_keywords in evaluation_items.items():
            if token in item_keywords:
                item_counts[item_category] += 1
    return item_counts

# 평가 항목 추출 결과
item_counts = extract_evaluation_items(tokens)
print("평가 항목 추출 결과:", item_counts)
```

```python
from transformers import BertTokenizer, BertForTokenClassification, BertTokenizerFast, Trainer, TrainingArguments
from datasets import load_dataset

# 데이터셋 로드 (일반적으로 많이 쓰이는 한국어 훈련모델: KorNLI, Korean NER)
ner_dataset = load_dataset("datasciathlete/corpus4everyone-klue-korean-NER") #Hugging Face Hub에서 datasete 찾기
#KorNLI(자연어추론):multi_nli(여러장르), snli(문장쌍), xnli(다국어)

# 토크나이저와 모델 로드
#ner_tokenizer = BertTokenizer.from_pretrained('bert-base-multilingual-cased')

# NER 모델과 토크나이저 로드
ner_tokenizer = BertTokenizerFast.from_pretrained('bert-base-multilingual-cased')
ner_model = BertForTokenClassification.from_pretrained('bert-base-multilingual-cased', num_labels=9)  # 9는 CoNLL-2003 태그 수
```

```python
def ner_tokenize_and_align_labels(examples):
    # Tokenizer 호출
    tokenized_inputs = ner_tokenizer(
        examples["tokens"],
        truncation=True,
        padding='max_length',
        is_split_into_words=True,
        return_offsets_mapping=True  # 'fast' 토크나이저에서만 지원됨
    )

    # 레이블 생성
    labels = []
    for i, label in enumerate(examples["ner_tags"]):
        # Offset mapping
        offset_mapping = tokenized_inputs.offset_mapping[i]
        label_ids = [-100] * len(offset_mapping)  # Initialize with -100

        # Align labels with tokens
        for j, (start, end) in enumerate(offset_mapping):
            if start != 0 and end != 0:  # Only set labels for non-special tokens
                word_id = tokenized_inputs.word_ids(batch_index=i)[j]
                if word_id is not None:
                    label_ids[j] = label[word_id]

        labels.append(label_ids)

    # 토큰화된 입력과 레이블을 반환
    tokenized_inputs["labels"] = labels
    return tokenized_inputs

# 데이터 전처리
ner_train_dataset = ner_dataset["train"].map(ner_tokenize_and_align_labels, batched=True)
ner_eval_dataset = ner_dataset["validation"].map(ner_tokenize_and_align_labels, batched=True)
```

```python
ner_training_args = TrainingArguments(
    output_dir="./ner_results",
    evaluation_strategy="epoch",
    learning_rate=5e-5, #원래는 2e-5
    per_device_train_batch_size=64, #원래는 16
    per_device_eval_batch_size=64, #원래는 16
    num_train_epochs=3,
    weight_decay=0.01,
    logging_dir="./ner_logs",
    fp16=True #mixed precision training 원래는 False
)

ner_trainer = Trainer(
    model=ner_model,
    args=ner_training_args,
    train_dataset=ner_train_dataset,
    eval_dataset=ner_eval_dataset,
)

# 모델 훈련
ner_trainer.train()
```
